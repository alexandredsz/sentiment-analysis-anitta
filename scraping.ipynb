```python
import snscrape.modules.twitter as sns

query = "(anitta) lang:pt"
tweets = []
max_size = 1000
for tweet in sns.TwitterSearchScraper(query).get_items():

    print(tweet.rawContent)
    print("===========")
    tweets.append(tweet.rawContent)
    if len(tweets) > max_size:
        break
```

```python
tweets
```

```python
import pandas as pd

df = pd.DataFrame({'text': tweets})

# saving the dataframe
df.to_csv('/content/drive/MyDrive/T2-IA/scrapping/tweets.csv')
```
